```python
# a test function to see if users with overlapping listing have similar taste
def similar(listingguid, df):
    same = df[df['listingGuid'].isin([listingguid])]
    users = same['visitor_guid'].tolist()
    visitors = df[df['visitor_guid'].isin(users)]
    # Same location?
    listing = visitors['listingGuid'].tolist()
    url = visitors['url'].tolist()
#     return listing, url
    return url
```
