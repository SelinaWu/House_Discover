```python
import pandas as pd
import numpy as np
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
%matplotlib inline 
sns.set(color_codes=True)
```

This document is a preliminary data analysis for the join_house_market_to_first_american file. The goal is just to understand the dataset better.

```python
# Read data
df = pd.read_csv('join_house_market_to_first_american_25025.csv')
```

```python
# See all the columns
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
# Check the data types
df.dtypes
```

```python
# see the number of rows and columns
df.shape
```

```python
# check if there are any duplicated rows
duplicate_rows_df = df[df.duplicated()]
print ('number of duplicate rows: ', duplicate_rows_df.shape)
```

```python
# see the counts to get an idea of the null values
df.count()
```

```python
# check the number of null values
print(df.isnull().sum())
```

```python
# drop rows that are all null values
df = df.dropna(how='all')
# drop columns that have more than 70% null values
df = df.dropna(thresh=df.shape[0]*0.3, axis=1, how='all')
# get an idea of how many columns are dropped
df.shape
```

```python
df.head(5)
```

```python
# check the number of null value again
print(df.isnull().sum())
```

```python
# remove any rows that have bedrooms or full baths as null as we think these are important features
df = df[df['bedrooms'].notnull()]
df = df[df['full_baths'].notnull()]
df.shape
```

```python
# check outliers
sns.boxplot(x=df['original_list_price'])
```

```python
# check outliers
sns.boxplot(x=df['current_list_price'])
```

```python
# check for outliers
sns.boxplot(x=df['current_list_price']-df['original_list_price'])
```

```python
# check for outiers
sns.boxplot(x=df['current_list_price']-df['sale_price'])
```

```python
# use heatmap to see the correlation between the variables
plt.figure(figsize=(20,10))
c= df.corr()
sns.heatmap(c,cmap='BrBG', annot=True)
c
```

We see that, as expected, the factors that are most important to the sale price is size (square foot), the bedroom number, and the full-bath number.

```python
# Plotting a scatter plot to see the relationships between sale price and full baths
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['sale_price'], df['full_baths'])
ax.set_xlabel('full_baths')
ax.set_ylabel('sale_price')
plt.show()
```

```python
# Plotting a scatter plot to see the relationships between sale price and bedrooms
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['sale_price'], df['bedrooms'])
ax.set_xlabel('bedrooms')
ax.set_ylabel('sale_price')
plt.show()
```

```python
# save the new dataset
df.to_csv(r'cleaned_listing_FirstAmerican.csv', index=False)
```
